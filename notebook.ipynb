### Cleaning Data
Memasuki proses terakhir dalam data wrangling yaitu pembersihan atau cleaning data. Pada tahap ini, kita akan membersihkan berbagai masalah yang telah teridentifikasi dalam proses assessing data.

Masalah berikutnya yang harus kita tangani ialah missing value pada kolom PM2.5, PM10, SO2, NO2, CO, O3, TEMP, PRESS, DEWP, RAIN, wd dan WSPM.

Terdapat tiga metode yang dapat digunakan untuk menangani *missing value* yaitu *dropping*, *imputation*, dan *interpolation*. Untuk menentukan metode mana yang akan digunakan, kita perlu melihat data yang mengandung *missing value* tersebut menggunakan teknik filtering seperti berikut.

####Menggunakan Metode Interpolation

- Untuk mengisi *missing value* pada kolom PM2.5, PM10, SO2, NO2, CO, O3 akan menggunakan metode interpolation, karena data nya berupa time series sehingga lebih cocok jika digunakan metode interpolation

```python
#Mengisi missing value pada kolom PM2.5, PM10, SO2, NO2, CO, O3, TEMP, PRES, DEWP, RAIN, WSPM

cols = ["PM2.5", "PM10", "SO2", "NO2", "CO", "O3", 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']

gabungan_df.interpolate(method='linear', inplace=True)

gabungan_df.isna().sum()
```

###Stasiun mana sajakah yang memiliki kadar polusi PM2.5, PM10, SO2, CO, O3 tertinggi dan terendah

```python
polusi_avg = gabungan_df.groupby("station").agg({
    "PM2.5": "mean",
    "PM10": "mean",
    "SO2": "mean",
    "CO": "mean",
    "O3": "mean",
}).reset_index()
polusi_avg
```

Diperoleh hasil rata-rata kadar tiap polusi udara dari setiap stasiun yang ada, mari kita permudah dalam membaca datanya

```python
for polusi in ["PM2.5", "PM10", "SO2", "CO", "O3"]:
  min_station = polusi_avg.loc[polusi_avg[polusi].idxmin()]
  max_station = polusi_avg.loc[polusi_avg[polusi].idxmax()]

  print(f"\nStasiun dengan {polusi} rata-rata TERENDAH: {min_station['station']} ({min_station[polusi]:.2f})")
  print(f"Stasiun dengan {polusi} rata-rata TERTINGGI: {max_station['station']} ({max_station[polusi]:.2f})")
```

**Interpretasi**
>Diperoleh bahwa stasiun Dingling merupakan stasiun yang memiliki kadar polusi udara rendah pada PM2.5, PM10, SO2, dan CO2. Namun juga menjadi stasiun dengan kadar polusi udara tertinggi karena kadar polusi udara O3 pada stasiun Dingling tertinggi

## Visualization & Explanatory Analysis

### Pertanyaan 1. Dari 13 Stasiun di Beijing, stasiun manakah yang memiliki kadar PM2.5 tertinggi dari tahun 2015-2017

```python
df_filtered = gabungan_df[(gabungan_df["year"] >= 2015) & (gabungan_df["year"] <= 2017)]

# Kelompokkan berdasarkan stasiun dan hitung rata-rata PM2.5
pm25_stasion = df_filtered.groupby("station")["PM2.5"].mean().reset_index()

# Plot barplot
plt.figure(figsize=(12, 6))
sns.barplot(x="station", y="PM2.5", data=pm25_stasion, palette="Reds_r")

# Tambahkan judul dan rotasi label stasiun
plt.xticks(rotation=90)
plt.xlabel("Station")
plt.ylabel("Rata-rata PM2.5 (g/m)")
plt.title("Rata-rata PM2.5 di Berbagai Stasiun (2015-2017)")

plt.show()
```

**Interpretasi**

Pertama kita perlu memasukan data yang sebelumnya kedalam variabel baru agar tidak mempengaruhi data sebelumnya, karena pada proses kali ini hanya akan menggunakan data pada 2 kolom saja yakni PM2.5 dan PM10

```python
data = pd.read_csv("gabungan_df.csv")

data_clustering = data.copy()
data_clustering
```

```python
# Drop kolom yang tidak dibutuhkan
data_clustering = data_clustering.drop(columns=["SO2","NO2","CO","O3","TEMP","PRES", "DEWP", "RAIN", "wd", "WSPM"])

data_clustering
```

####Untuk data pada kolom PM2.5 akan dimasukkan dalam fiturx dan data pada kolom PM10 akan dimasukkan ke dalam kolom fitury

```python
fiturx = data["PM2.5"].values
fitury = data["PM10"].values
```

#####Akan digunakan nilai quartile 1 dan quartile 2 sebagai batas

```python
q1=np.percentile(fiturx,25)
q2=np.percentile(fiturx,75)
```

#####Membuat batasan data menjadi 3 grup yakni rendah, standar dan tinggi

```python
polusi_rendah = data[data["PM2.5"] < q1]
polusi_standar = data [(data["PM2.5"]>=q1) & (data["PM2.5"]<q2) ]
polusi_tinggi = data[data["PM2.5"] >= q2]
```

#####Menghitung centroid untuk masing-masing cluster menggunakan rata-rata dari data PM2.5 dan PM10

```python
centroid_rendah = [polusi_rendah["PM2.5"].mean(), polusi_rendah["PM10"].mean()]
centroid_standar = [polusi_standar["PM2.5"].mean(),polusi_standar["PM10"].mean()]
centroid_tinggi = [polusi_tinggi["PM2.5"].mean(), polusi_tinggi["PM10"].mean()]
```

##### Menentukan cluster berdasarkan kedekatan dengan centroid, dengan menghitung jarak ke setiap centroid, kemudian hasil dari perhitungan tersebut akan disesuaikan dengan kondisi if else yang memenuhi

```python
clusters = []
for i in range(len(data)):
    point = [data["PM2.5"].iloc[i], data["PM10"].iloc[i]]

    # Hitung jarak ke setiap centroid
    dist_rendah = np.linalg.norm(np.array(point) - np.array(centroid_rendah))
    dist_standar = np.linalg.norm(np.array(point) - np.array(centroid_standar))
    dist_tinggi = np.linalg.norm(np.array(point) - np.array(centroid_tinggi))
    # Menentukan cluster berdasarkan jarak terdekat
    if dist_rendah < dist_standar and dist_rendah < dist_tinggi:
        cluster = 1
    elif dist_standar < dist_rendah and dist_standar < dist_tinggi:
        cluster = 2
    else:
        cluster = 3

    clusters.append(cluster)
```

####Selanjutnya data disimpan dengan label cluster masing-masing

```python
data_with_clusters = data.copy()
data_with_clusters['cluster'] = clusters

# Menyimpan DataFrame untuk setiap cluster ke dalam file CSV terpisah dan mengunduhnya
for cluster_label in range(1, 4):
    cluster_data = data_with_clusters[data_with_clusters['cluster'] == cluster_label]
    filename = f'data_cluster_{cluster_label}.csv'
    cluster_data.to_csv(filename, index=False)

# Tampilkan hasil clustering
print("\nData dengan label cluster:")
print(data_with_clusters)

# Tampilkan centroid untuk setiap cluster
print(f"\nCentroid untuk cluster rendah: {centroid_rendah}")
print(f"Centroid untuk cluster standar: {centroid_standar}")
print(f"Centroid untuk cluster tinggi: {centroid_tinggi}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_with_clusters[data_with_clusters['cluster'] == 1]['PM2.5'],
            data_with_clusters[data_with_clusters['cluster'] == 1]['PM10'],
            color='red', label='Cluster 1')

plt.scatter(data_with_clusters[data_with_clusters['cluster'] == 2]['PM2.5'],
            data_with_clusters[data_with_clusters['cluster'] == 2]['PM10'],
            color='blue', label='Cluster 2')

plt.scatter(data_with_clusters[data_with_clusters['cluster'] == 3]['PM2.5'],
            data_with_clusters[data_with_clusters['cluster'] == 3]['PM10'],
            color='green', label='Cluster 3')

plt.scatter([centroid_rendah[0], centroid_tinggi[0], centroid_standar[0]],
            [centroid_rendah[1], centroid_tinggi[1], centroid_standar[1]],
            color='black', marker='X', s=200, label='Centroids')

plt.xlabel("PM2.5")
plt.ylabel("PM10")
plt.title("Clustering Stasiun Kadar Polusi Rendah, Standar dan Tinggi")
plt.legend()
plt.grid(True)
plt.show()
```
